```python
#I- Exploratory Data Analysis 

# 1. Import Required Libraries

import pandas as pd
import numpy as np

# 2. Load the Dataset

file_path = r"C:\Users\Abraham\Documents\Dell\DSTI\Projets Data\hr-retail-dashboard\Data\MFGEmployees4.csv"
df = pd.read_csv(file_path)
```

```python
# 3. Preview the First Rows

print("\nFirst five rows of the dataset:\n")
display(df.head())
```

```python
# 4. Dataset Structure and Missing Values
print("\nDataset informations:\n")
df.info()

print("\nCount of missing values per column:\n")
print(df.isnull().sum())

# 5. Descriptive Statistics
print("\nDescriptive statistics for all columns:\n")
display(df.describe(include='all'))
```

```python
# 6. Observations and Initial Data Quality Assessment

display(df.describe(include='all'))

observations = """
Observations from data inspection:

1. The values in the 'Age' column are stored as floats. For clarity and consistency, they should be rounded and converted to integers.
    The minimum value for 'Age' is 3.5 years, which is unrealistic for employees and likely represents a data entry error.

2. The minimum 'LengthService' is 0.012 years (~4.4 days) [0.012 year * 365 days].
   While possible for new hires, this should be checked against business rules.

3. Some employees have 'AbsentHours' equal to 0.
   Should be cross-checked (especially for long-service employees).

4. Check for duplicate records using 'EmployeeNumber'.
   Notable name frequencies: "Johnson" (106), "James" (182).

5. Check for hidden missing values (empty strings, placeholders).

6. Investigate high absenteeism (>=150 hours).

7. Investigate long service years (>=40 years).
"""
print(observations)
```

```python
#II - Ivestigations

# 1. Employees with Age < 14
print("\nEmployees with Age < 14:\n")
print(df[df['Age'] < 14])

Comment_1 = """
These observations refer to individuals with an age below 14, 
which is below the legal working age in Canada. These entries are likely data errors 
or outliers, and will therefore be removed from the dataset.
"""
print(Comment_1)
```

```python
# 2. Employees with LengthService < 1 month (~0.083 years) [1 month /12 months]
print("\nEmployees with LengthService < 0.083 years:\n")
print(df[df['LengthService'] < 0.083])

Comment_2 = """
These 2 lines are refer to individuals who have spent less than 1 month inside the organization, therefore they are not relevant for the analysis and will be removed from the dataset.
"""
print(Comment_2)
```

```python
# 3. Employees with zero AbsentHours and LengthService > 5 years
zero_abs_long_service = df[(df['AbsentHours'] == 0) & (df['LengthService'] > 5)]
print("\nEmployees with zero AbsentHours and LengthService > 5 years:\n")
print(zero_abs_long_service)

Comment_3 = f"""
There are {len(zero_abs_long_service)} employees with more than 5 years of service and zero recorded absent hours.
While this could indicate perfect attendance, it is uncommon over such a long period.
However, we will not remove their records automatically, as they may also be valid.
"""
print(Comment_3)
```

```python
# 4. Duplicate EmployeeNumber check
print("\nDuplicate EmployeeNumber records:\n")
print(df[df.duplicated('EmployeeNumber', keep= False)])

Comment_4 = """
Through this output we can confirm that there are no duplicate EmployeeNumber entries in the dataset.
Each EmployeeNumber appears only once, which ensures data integrity for this key identifier.
"""
print(Comment_4)
```

```python
# 5. Check for hidden missing values ( (empty strings, 'N/A', 'Unknown', etc.)

print("\nCount of possible hidden missing values per column:\n")

# Liste des valeurs à vérifier
missing_placeholders = ['', 'N/A', 'NA', 'n.a', 'Unknown', '...', '.', '/', ';', ',']

hidden_missing = False
for col in df.columns:
    if df[col].dtype == object:
        counts = {val: (df[col] == val).sum() for val in missing_placeholders}
        total = sum(counts.values())
        if total > 0:
            hidden_missing = True
            details = ', '.join([f"{repr(k)}={v}" for k, v in counts.items() if v > 0])
            print(f"{col}: {details}")

if not hidden_missing:
    print("No hidden missing values found in any column.")

Comment_5 = """
Through this output, we confirm that there are no possible hidden missing values (such as empty strings, N/A, n.a, Unknown, ..., ., /, ;, ',') in the dataset.
"""
print(Comment_5)
```

```python
# 6. Employees with AbsentHours >= 150
high_absenteeism = df[df['AbsentHours'] >= 150]
print(f"\nEmployees with AbsentHours >= 150: {len(high_absenteeism)} found\n")
print(high_absenteeism)

Comment_6 = f"""
There are {len(high_absenteeism)} employees with 150 or more recorded absent hours.
This represents approximately {len(high_absenteeism)/len(df)*100:.0f}% of all entries in the dataset.
Given this proportion, we will not remove these records, as high absenteeism may be valid for certain cases.
"""
print(Comment_6)
```

```python
# 7. Employees with LengthService >= 40 years
print("\nEmployees with LengthService >= 40 years:\n")
print(df[df['LengthService'] >= 40])


Comment_7 = """
There is only 1 employee with 40 or more years of service.
Given the uniqueness and plausibility of such long tenure, we will not remove this record from the dataset.
"""
print(Comment_7)
```

```python
#III - Data Cleaning 

df_cleaned = df.copy()

print("\nApplying cleaning recommendations...\n")
```

```python
## Recommendation 1: Delete employees under 14 years of age and convert Age to integer

initial_rows = len(df)
df_cleaned = df_cleaned[df_cleaned['Age'] >= 14]

df_cleaned['Age'] = df_cleaned['Age'].round().astype(int)

## Recommendation 2: Delete employees less than 1 month old

df_cleaned = df_cleaned[df_cleaned['LengthService'] >= 0.083]

# Calculates cleaning statistics

final_rows = len(df_cleaned)
rows_removed = initial_rows - final_rows
```

```python
#IV - Save and report

# 1. Save cleaned dataset

output_path = r"C:\Users\Abraham\Documents\Dell\DSTI\Projets Data\hr-retail-dashboard\Data\MFGEmployees4_cleaned.csv"
df_cleaned.to_csv(output_path, index=False)

# 2. Generate cleanup report

cleaning_report = f"""
Data Cleaning Report :

Original dataset size : {len(df)} rows
Cleaned dataset size  : {final_rows} rows
Rows removed          : {rows_removed} rows

Cleaning actions applied:
1. Age < 14              : {len(df[df['Age'] < 14])} rows removed
2. LengthService < 1 month : {len(df[df['LengthService'] < 0.083])} rows removed

Dataset saved to:
{output_path}
"""

print(cleaning_report)

# 3. Final check

print("\nFinal check after cleaning:\n")
display(df_cleaned.describe(include='all'))
```
